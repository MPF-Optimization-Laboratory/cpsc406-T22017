```julia
using Polynomials
```

```julia
p = poly( [-4, -2, -1, 1, 3] )
```

```julia
xlim = (-4.1, 3.1)
plot( t->p(t), xlim... )
r = roots(p)
scatter!(r, zeros(r))
```

```julia
function gen_data(p, m, )
   srand(0)
   t = randn(m)
   noise = randn(m); noise *= /norm(noise)
   b = p(t) + noise
   return (t,b)
end
```

```julia
# Generate random data
t = b = []
m = 100 # no. of samples
 = 20 # % of noise
t, b = gen_data(p, m, );
scatter( t, b, xlim=xlim, ylim=(-100,60) )
```

```julia
# Compute polynomial fit
d = 3 # degree of fit
pp = polyfit(t,b,d)
scatter( t, b, xlim=xlim, ylim=(-100,60) )
plot!( t->pp(t), xlim... )
```

```julia
d = 5
A = [ t[i]^k for i=1:m, k=0:d ]
x = A \ b # == inv(A)*b
pp = Poly(x)
plot!( t->pp(t), xlim..., ylim=(-100,60) )
```

```julia
x1 = A \ b
x2 = (A'A) \ (A'b)
norm(x1 - x2)
```
